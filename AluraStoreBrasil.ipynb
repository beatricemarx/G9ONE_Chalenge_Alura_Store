### Importação dos dados

# 4. Produtos Mais e Menos Vendidos

```python
# Neste passo, deve identificar quais produtos foram os mais vendidos e os menos vendidos em cada loja.
# Ao final, visualize os resultados de forma que fique claro quais produtos se destacaram nas vendas de cada loja.


produtos_loja_1 = loja.groupby(['Filial', 'Categoria do Produto', 'Produto'])['Produto'].count().sort_values(ascending=False).to_frame(name= 'Quantidade de Vendas')

produtos_loja_2 = loja2.groupby(['Filial', 'Categoria do Produto', 'Produto'])['Produto'].count().sort_values(ascending=False).to_frame(name= 'Quantidade de Vendas')

produtos_loja_3 = loja3.groupby(['Filial', 'Categoria do Produto', 'Produto'])['Produto'].count().sort_values(ascending=False).to_frame(name= 'Quantidade de Vendas')

produtos_loja_4 = loja4.groupby(['Filial', 'Categoria do Produto', 'Produto'])['Produto'].count().sort_values(ascending=False).to_frame(name= 'Quantidade de Vendas')


# Filtrar o PRODUTO MAIS VENDIDO por Filial
mais_vendido_loja1 = produtos_loja_1.head(1)

mais_vendido_loja2 = produtos_loja_2.head(1)

mais_vendido_loja3 = produtos_loja_3.head(1)

mais_vendido_loja4 = produtos_loja_4.head(1)


# 4. Filtrar o PRODUTO MENOS VENDIDO por Filial
menos_vendido_loja1 = produtos_loja_1.tail(1)

menos_vendido_loja2 = produtos_loja_2.tail(1)

menos_vendido_loja3 = produtos_loja_3.tail(1)

menos_vendido_loja4 = produtos_loja_4.tail(1)


mais_vendido = pd.concat([mais_vendido_loja1, mais_vendido_loja2, mais_vendido_loja3, mais_vendido_loja4])

menos_vendido = pd.concat([menos_vendido_loja1, menos_vendido_loja2, menos_vendido_loja3, menos_vendido_loja4])

mais_vendido
```

```python
# Valor total de Vendas por Filial para as Categorias Eletrodomesticos e Eletronicos

import matplotlib.pyplot as plt
import numpy as np

# Prepare os dados para o gráfico de barras empilhadas
# Desempilhe 'Categoria do Produto' para tê-los como colunas
sales_pivot = df_comp_faturamento_categoria_filial_filtrado.unstack(level='Categoria do Produto')

# Extraia os dados para plotagem
filiais = sales_pivot.index
eletrodomesticos_sales = sales_pivot[('Faturamento Total', 'eletrodomesticos')]
eletronicos_sales = sales_pivot[('Faturamento Total', 'eletronicos')]

# Crie o gráfico de barras
fig, ax = plt.subplots(figsize=(5, 3))

# Plote as vendas de 'eletrodomesticos'
ax.bar(filiais, eletrodomesticos_sales, label='Eletrodomésticos', color='skyblue')

# Empilhe as vendas de 'eletronicos' em cima das vendas de 'eletrodomesticos'
ax.bar(filiais, eletronicos_sales, bottom=eletrodomesticos_sales, label='Eletrônicos', color='lightcoral')

# Adicione rótulos e título
ax.set_ylabel('Vendas')
ax.set_title('Vendas por Filial e Categoria (Eletrodomésticos e Eletrônicos)')
ax.legend()
plt.xticks(rotation=45, ha='right') # Rotaciona os rótulos do eixo x para melhor legibilidade
plt.tight_layout()
plt.show()
```
